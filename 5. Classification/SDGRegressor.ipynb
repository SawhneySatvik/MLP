## Logistic Regression

```python
bin_sgd_clf = SGDClassifier(loss='log_loss',
                            penalty='l2',
                            max_iter=1,
                            warm_start=True,
                            eta0=0.01,
                            alpha=0,
                            learning_rate='constant',
                            random_state=1729
                            )
Loss = []
iterations = 100
for i in range(iterations):
    bin_sgd_clf.fit(x_train, y_train_0)
    y_pred = bin_sgd_clf.predict_proba(x_train)
    Loss.append(log_loss(y_train_0, y_pred))
```

```python
plt.figure()
plt.plot(np.arange(iterations), Loss)
plt.grid(True)
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.show()
```

```python
print("Train Accuracy: ", bin_sgd_clf.score(x_train, y_train_0))
print('Test Accuracy: ', bin_sgd_clf.score(x_test, y_test_0))
```
